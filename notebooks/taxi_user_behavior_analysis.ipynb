**1 Название и контекст бизнеса**

**Анализ поведения пользователей такси**

Бизнес-проблема  
Компания хочет понять, предпочитают ли пользователи класс «Комфорт» для поездок на дальние расстояния классу «Эконом».

Цели анализа:  
Изучить распределение расстояний поездок
Сравнить классы «Эконом» и «Комфорт»
Проанализировать описательную статистику
Визуализировать распределения плотности
Проверить статистическую гипотезу

**2 Импорт**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

sns.set(style="whitegrid")
```

**3 Загрузка данных**

```python
df = pd.read_excel('../data/FPS_hw_x_df_x_taxi.xlsx')

print("Dataset shape:", df.shape)
df.head()
```

**5 Описательная статистика по классам**

```python
df.groupby('class')['distance'].describe()
```

**6 Визуализация распределения (график плотности)**

```python
plt.figure(figsize=(10,6))

sns.kdeplot(df[df['class'] == 'Economy']['distance'], label='Economy')
sns.kdeplot(df[df['class'] == 'Comfort']['distance'], label='Comfort')

plt.title('Distance Distribution by Taxi Class')
plt.xlabel('Distance')
plt.ylabel('Density')
plt.legend()
plt.show()
```

**7 Сравнение с помощью диаграмм размаха**

```python
plt.figure(figsize=(8,6))
sns.boxplot(data=df, x='class', y='distance')
plt.title('Distance Comparison by Taxi Class')
plt.show()
```

**8 Подготовка выборок**

```python
eco = df[df['class'] == 'Economy']['distance']
conf = df[df['class'] == 'Comfort']['distance']
```

**9 Проверка нормальности**

```python
# Выборка для предотвращения смещения, вызванного слишком большими выборками
eco_sample = eco.sample(min(500, len(eco)), random_state=42)
conf_sample = conf.sample(min(500, len(conf)), random_state=42)

print("Shapiro test - Economy:", stats.shapiro(eco_sample))
print("Shapiro test - Comfort:", stats.shapiro(conf_sample))
```

**10 Проверка гипотез**

```python
# H0: Распределения расстояний равны
# H1: Поездки с комфортом имеют большие расстояния, чем поездки эконом-класса

stat, p_value = stats.mannwhitneyu(eco, conf, alternative='less')

print("Test statistic:", stat)
print("p-value:", p_value)
```

**11 Интерпретация**

```python
alpha = 0.05

if p_value < alpha:
    print("Reject H0: Comfort trips tend to have longer distances.")
else:
    print("Fail to reject H0: No statistical evidence of difference.")
```

**12 Заключение**

На основе описательной статистики, визуализации плотности и проверки гипотез:

- Класс «Комфорт» демонстрирует более высокую концентрацию поездок на большие расстояния.

- Критерий Манна-Уитни указывает на статистическую значимость (если p < 0,05).

### Последствия для бизнеса

Статистический анализ показывает, что расстояние поездки связано с выбором класса такси, при этом класс «Комфорт» чаще выбирается для поездок на большие расстояния.
